```python
california_housing_train.csv 파일은 1990년 캘리포니아 인구 조사 데이터를 기반으로 만들어진 캘리포니아 주택 가격 예측 모델 훈련용 데이터셋입니다. 🏠

이 데이터는 머신러닝, 특히 회귀(Regression) 모델을 학습하고 평가하는 데 아주 널리 사용됩니다. 각 행은 캘리포니아의 한 **블록 그룹(block group)**에 대한 정보를 담고 있습니다. 블록 그룹은 미국 인구조사에서 사용하는 최소 지리 단위로, 보통 600명에서 3,000명의 인구를 포함합니다.

데이터셋의 목표는 주어진 여러 특성(feature)들을 사용해 해당 지역의 **중간 주택 가격(median_house_value)**을 예측하는 것입니다.
```

```python
컬럼별 상세 설명
컬럼명	설명	데이터 타입
longitude	경도	수치형 (Numeric)
latitude	위도	수치형 (Numeric)
housing_median_age	주택 나이의 중간값	수치형 (Numeric)
total_rooms	전체 방 개수	수치형 (Numeric)
total_bedrooms	전체 침실 개수	수치형 (Numeric)
population	총인구	수치형 (Numeric)
households	총가구 수	수치형 (Numeric)
median_income	소득 중간값 (단위: 만 달러, $10,000s)	수치형 (Numeric)
median_house_value	주택 가격 중간값 (단위: 미국 달러)	수치형 (Numeric)
```

```python
💡 주요 특징 및 활용
타겟 변수: median_house_value가 예측해야 할 목표값(Target)입니다.

지리적 정보: longitude와 latitude 컬럼은 주택의 위치 정보를 나타내며, 이를 통해 해안가나 특정 도시와 같은 위치적 요인이 주택 가격에 미치는 영향을 분석할 수 있습니다.

단위 확인 필요: median_income은 만 달러 단위로 스케일링 되어 있습니다. 예를 들어, 값이 3이면 실제 소득은 $30,000를 의미합니다.

파생 변수 생성 가능: total_rooms / households (가구당 방 개수), total_bedrooms / total_rooms (방 대비 침실 비율), population / households (가구당 인원) 등과 같이 기존 컬럼을 조합하여 새로운 의미 있는 특성(파생 변수)을 만들어 모델의 성능을 높이는 데 활용할 수 있습니다.

결측치: 이 데이터셋의 total_bedrooms 컬럼에는 종종 결측치(missing value)가 포함되어 있어, 데이터 전처리 과정에서 이를 처리하는 과정이 필요합니다.
```

```python
import matplotlib.pyplot as plt

# Select columns for visualization
distribution_cols = ['median_house_value', 'median_income', 'population', 'housing_median_age']
spatial_cols = ['longitude', 'latitude']

# Create histograms for distribution columns
plt.figure(figsize=(15, 10))
for i, col in enumerate(distribution_cols):
    plt.subplot(2, 2, i + 1)
    plt.hist(df_sales[col], bins=50, color='skyblue', edgecolor='black')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {col}')

# Create a scatter plot for spatial distribution
plt.figure(figsize=(10, 8))
plt.scatter(df_sales['longitude'], df_sales['latitude'], alpha=0.5,
            c=df_sales['median_house_value'], cmap='viridis', s=df_sales['population']/100)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of Housing Data (Color by Median House Value, Size by Population)')
plt.colorbar(label='Median House Value')

plt.tight_layout()
plt.show()
```

```python
입력 프롬프트

경찰서 데이터를 구별로 통합하여 구별 범죄 현황 및 특성을 분석하고 싶어. 구글맵API를 사용해서 각 경찰서가 어떤 구에 속하는지 처리하는 코드 작성해줘
```

```python
위 코드에서 입력규칙은
중부서 -> 서울중부경찰서,
서대문서-> 서울서대문경찰서,
종로서->서울종로경찰서 와 같아.
마지막 '서'글자를 제외하고 앞과 뒤에 '서울'과 '경찰서'를 붙이도록 코드 다시 작성
```

```python
import googlemaps

def get_district_from_police_station(police_station_name):
    """
    경찰서 이름을 입력받아 Google Maps Geocoding API를 사용하여 해당 경찰서가 속한 '구' 이름을 반환합니다.
    """
    try:
        # Apply the user's rule for formatting the police station name
        formatted_name = '서울' + police_station_name[:-1] + '경찰서'

        # Geocoding search with the formatted name
        geocode_result = gmaps.geocode(formatted_name, language='ko')

        if geocode_result:
            # Extract administrative area information from the result
            for component in geocode_result[0]['address_components']:
                if 'sublocality_level_1' in component['types']:
                    return component['long_name']
                elif 'locality' in component['types']:
                     return component['long_name'] # Fallback for cases where sublocality_level_1 is not available
            return None # Return None if '구' information is not found
        else:
            return None # Return None if no geocoding result is found

    except Exception as e:
        print(f"Error processing {police_station_name}: {e}")
        return None
```

```python
# Apply the function to the '관서명' column to get the district for each police station
crime_anal_police['구'] = crime_anal_police['관서명'].apply(get_district_from_police_station)

# Display the dataframe with the new '구' column
display(crime_anal_police[['관서명', '구']].head())
```

```python
import pandas as pd

# Load the CCTV data
cctv_url = 'https://raw.githubusercontent.com/jiminxchris/ml-practice/main/data/CCTV_result.csv'
df_cctv = pd.read_csv(cctv_url)
display(df_cctv.head())

# Load the population data
population_url = 'https://github.com/jiminxchris/ml-practice/blob/main/data/population_in_Seoul.xls?raw=true'
df_population = pd.read_excel(population_url, header=2, usecols='B, D, G, J, N') # Assuming relevant columns and header row
display(df_population.head())

# Rename columns for merging
df_cctv.rename(columns={'구별': '구'}, inplace=True)
# Corrected renaming based on df_population.head() output
df_population.rename(columns={'자치구': '구', '계': '인구수', '계.1': '한국인_인구수', '계.2': '외국인_인구수', '65세이상고령자': '고령자_인구수'}, inplace=True)


# Merge crime data with CCTV data
crime_anal_merged = pd.merge(crime_anal_police, df_cctv[['구', '소계']], on='구', how='left')
crime_anal_merged.rename(columns={'소계': 'CCTV 총계'}, inplace=True)
display(crime_anal_merged.head())

# Merge the combined data with population data
crime_anal_merged = pd.merge(crime_anal_merged, df_population[['구', '인구수', '한국인_인구수', '외국인_인구수', '고령자_인구수']], on='구', how='left')
display(crime_anal_merged.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate total crime occurrence per district (summing across police stations within the same district if any)
# First, group by '구' and sum the occurrence columns
crime_occurrence_cols = ['살인 발생', '강도 발생', '강간 발생', '절도 발생', '폭력 발생']
district_crime_summary = crime_anal_merged.groupby('구')[crime_occurrence_cols].sum().reset_index()

# Calculate total crime per district
district_crime_summary['총 범죄 발생'] = district_crime_summary[crime_occurrence_cols].sum(axis=1)

# Merge with population and CCTV data at the district level
# Need to handle cases where multiple police stations are in one district (like 중구, 종로구, 강남구, 서초구, 은평구, 성북구)
# For population and CCTV, we need the district-level values, which are already in df_population and df_cctv
district_info = pd.merge(df_population[['구', '인구수']], df_cctv[['구', '소계']], on='구', how='left')
district_info.rename(columns={'소계': 'CCTV 총계'}, inplace=True)

# Merge the crime summary with district info
crime_district_viz = pd.merge(district_crime_summary, district_info, on='구', how='left')

# Remove the row with missing '구' data (금천서) if it wasn't merged with population/CCTV data
crime_district_viz.dropna(subset=['구'], inplace=True)

# Create scatter plot: Total Crime vs. Population
plt.figure(figsize=(10, 6))
sns.scatterplot(data=crime_district_viz, x='인구수', y='총 범죄 발생', hue='구', s=100)
plt.title('인구수와 총 범죄 발생 건수 관계')
plt.xlabel('인구수')
plt.ylabel('총 범죄 발생 건수')
plt.legend(title='구', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

# Create scatter plot: Total Crime vs. CCTV Count
plt.figure(figsize=(10, 6))
sns.scatterplot(data=crime_district_viz, x='CCTV 총계', y='총 범죄 발생', hue='구', s=100)
plt.title('CCTV 총계와 총 범죄 발생 건수 관계')
plt.xlabel('CCTV 총계')
plt.ylabel('총 범죄 발생 건수')
plt.legend(title='구', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
moonphase 컬럼을 'first quarter'-> 0.5, 'full moon'-> 1, 'third quarter'-> 0.5, 'new moon' ->0로 변경해줘
```

```python
# Define the mapping from moon phase names to numerical values
moonphase_mapping = {
    'first quarter': 0.5,
    'full moon': 1,
    'third quarter': 0.5,
    'new moon': 0
}

# Apply the mapping to the 'moonphase' column
df_moonphases['moonphase_numeric'] = df_moonphases['moonphase'].map(moonphase_mapping)

# Display the relevant columns to verify the change
display(df_moonphases[['moonphase', 'moonphase_numeric']].head())
display(df_moonphases[['moonphase', 'moonphase_numeric']].tail())
```

```python
# Fill missing values in 'moonphase_numeric' with the previous valid observation
df_moonphases['moonphase_numeric'] = df_moonphases['moonphase_numeric'].fillna(method='ffill')

# Display the relevant columns to verify the change
display(df_moonphases[['moonphase', 'moonphase_numeric']].head(20))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the realistic sales data
url = 'https://raw.githubusercontent.com/jiminxchris/ml-practice/main/data/realistic_sales_data.csv'
df_sales_realistic = pd.read_csv(url)

# Display the first few rows to understand the data structure
display(df_sales_realistic.head())
```

```python
# Convert the 'date' column to datetime objects
df_sales_realistic['date'] = pd.to_datetime(df_sales_realistic['date'])

# Extract the month from the 'date' and create a 'month' column
df_sales_realistic['month'] = df_sales_realistic['date'].dt.month

# Group by month and sum the sales
monthly_sales = df_sales_realistic.groupby('month')['sales'].sum().reset_index()

# Map month numbers to names for better visualization
month_names = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
}
monthly_sales['month_name'] = monthly_sales['month'].map(month_names)

# Create a bar chart of monthly sales
plt.figure(figsize=(8, 4))
plt.bar(monthly_sales['month_name'], monthly_sales['sales'], color='lightcoral')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Volume (Realistic Data)')
plt.tight_layout()
plt.show()
```

```python
# Extract the year from the 'date' column and create a 'year' column
df_sales_realistic['year'] = df_sales_realistic['date'].dt.year

# Group by year and month and sum the sales
yearly_monthly_sales = df_sales_realistic.groupby(['year', 'month'])['sales'].sum().reset_index()

# Display the result
display(yearly_monthly_sales.head())
display(yearly_monthly_sales.tail())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Combine year and month for plotting on a continuous time axis
yearly_monthly_sales['year_month'] = pd.to_datetime(yearly_monthly_sales['year'].astype(str) + '-' + yearly_monthly_sales['month'].astype(str))

plt.figure(figsize=(8, 4))
sns.lineplot(data=yearly_monthly_sales, x='year_month', y='sales')

plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Trend Over Time (Realistic Data)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate monthly statistics (average, max, min sales)
monthly_stats = df_sales_realistic.groupby(df_sales_realistic['date'].dt.month)['sales'].agg(['mean', 'max', 'min']).reset_index()
monthly_stats['month'] = monthly_stats['date'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
monthly_stats = monthly_stats.set_index('month').reindex([
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
]).reset_index()

# Calculate yearly statistics (average, max, min sales)
yearly_stats = df_sales_realistic.groupby(df_sales_realistic['date'].dt.year)['sales'].agg(['mean', 'max', 'min']).reset_index()
yearly_stats.rename(columns={'date': 'year'}, inplace=True)

# Visualize monthly statistics
plt.figure(figsize=(8, 4))
plt.plot(monthly_stats['month'], monthly_stats['mean'], marker='o', linestyle='-', label='Average Sales')
plt.plot(monthly_stats['month'], monthly_stats['max'], marker='o', linestyle='-', label='Maximum Sales')
plt.plot(monthly_stats['month'], monthly_stats['min'], marker='o', linestyle='-', label='Minimum Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Statistics (Average, Max, Min)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Visualize yearly statistics
plt.figure(figsize=(8, 4))
plt.plot(yearly_stats['year'], yearly_stats['mean'], marker='o', linestyle='-', label='Average Sales')
plt.plot(yearly_stats['year'], yearly_stats['max'], marker='o', linestyle='-', label='Maximum Sales')
plt.plot(yearly_stats['year'], yearly_stats['min'], marker='o', linestyle='-', label='Minimum Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Yearly Sales Statistics (Average, Max, Min)')
plt.legend()
plt.grid(True)
plt.xticks(yearly_stats['year']) # Ensure all years are shown on the x-axis
plt.tight_layout()
plt.show()
```
